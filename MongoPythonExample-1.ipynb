In real world, all the food ordering companies like Zomato, FoodPanda, Swigy etc are using NoSQL database for storing such data.So in this example we are going to generate some online restaurant review data and insert to the MongoDB table. Here each record is going to have 3 different fields "name", "rating" and cuisine. 

```python
from pymongo import MongoClient
from random import randint
#Step 1: Connect to MongoDB - Note: Change connection string as needed
client = MongoClient(port=27017)
db=client.business
#Step 2: Create sample data
names = ['Kitchen Pvt Ltd','HomeFood Pvt Ltd','State Pvt Ltd', 'Tastey Pvt Ltd', 'Big Pvt Ltd','City Pvt Ltd','Ghar ka khana Pvt Ltd', 'Pizza Pvt Ltd','Home Mane Pvt Ltd', 'Salty Pvt Ltd','Sandwich Pvt Ltd','Lazy Pvt Ltd', 'Fun Pvt Ltd']
company_cuisine = ['Pizza', 'Indian', 'Fast Food', 'Italian', 'Mexican', 'American', 'Chinese', 'Vegetarian']
for x in range(1, 501):
    business = {
        'name' : names[randint(0, (len(names)-1))],
        'rating' : randint(1, 5),
        'cuisine' : company_cuisine[randint(0, (len(company_cuisine)-1))] 
    }
    #Step 3: Insert business object directly into MongoDB via isnert_one
    result=db.reviews.insert_one(business)
    #Step 4: Print to the console the ObjectID of the new document
    print('Created {0} of 500 as {1}'.format(x,result.inserted_id))
#Step 5: Tell us that you are done
print('finished creating 500 business reviews')
```

let's say company want to know those customer reviews where rating provided was very less such as 1 then company need to run below code to find out those food order details.

```python
onestarbookings = db.reviews.find({'rating': 1})
for  foodorder in onestarbookings:
    print(foodorder)
    
    
```

Let say one of restaurant whose name was "HomeFood Pvt Ltd" has changed its name to "FoodMadeInHome Pvt Ltd". Now we need to update those records which have neen written into MongoDB to replace "HomeFood" to "FoodMadeInHome". Write a code to do this. 

```python
onestarbookings = db.reviews.find({"name":"HomeFood Pvt Ltd"})
for  foodorder in onestarbookings:
     foodorder['name']="FoodMadeInHome Pvt Ltd"
     print(foodorder)   
```
